## Data Structures

### Metric Encapsulation

The `EssayMetrics` [dataclass](https://docs.python.org/3/library/dataclasses.html) serves as a structured container for stylometric features. By decoupling data storage from logic, we ensure the pipeline remains 'extensible' – allowing for the addition of new metrics (e.g. Flesch-Kincaid readability) without refactoring the orchestrator.

````python
@dataclass
class EssayMetrics:
  '''All computed metrics for a single essay text.'''
  label: str

  # 1. Syntactic & punctuation
  colon_count: int = 0
  semicolon_count: int = 0
  colon_semicolon_ratio: float = 0.0   # colons / (semicolons + 1)
  sentence_lengths: list[int] = field(default_factory=list)
  mean_sentence_length: float = 0.0
  sentence_length_std: float = 0.0     # 'burstiness' proxy
  hedge_frequency: float = 0.0         # hedges per 100 words
  discourse_marker_frequency: float = 0.0

  # 2. Lexical
  ttr: float = 0.0                     # type–token ratio
  gptism_frequency: float = 0.0        # GPT-ism words per 100 words
  gptism_counts: dict[str, int] = field(default_factory=dict)

  # 3. Structural
  paragraph_lengths: list[int] = field(default_factory=list)
  paragraph_length_std: float = 0.0
  signpost_frequency: float = 0.0
  nuance_trap_frequency: float = 0.0   # 'while X ... Y' constructions per 100 words
  intro_outro_similarity: float = 0.0  # cosine sim between first & last paragraph

  # 4. Sophisticated
  mean_intra_paragraph_similarity: float = 0.0  # semantic redundancy
  intra_paragraph_similarities: list[float] = field(default_factory=list)
````

## Section 1 — Text Preprocessing

### Stylometric Foundations

Preprocessing targets 'function words' and 'GPT-isms'. Stylometry – the statistical study of linguistic style – suggests that frequent, subconscious word choices (like discourse markers) are more reliable indicators of authorship than thematic content. We use regular expressions (regex) to anchor counts to word boundaries, ensuring precise frequency measurements.

````python
# -- Vocabulary lists ----------------------------------------------------------

# Define common hedging phrases that reduce the assertiveness of a statement
HEDGE_WORDS: frozenset[str] = frozenset({
  'arguably', 'potentially', 'largely', 'generally', 'typically',
  'usually', 'often', 'somewhat', 'perhaps', 'possibly', 'likely',
  'seemingly', 'apparently', 'presumably', 'to some extent',
  'it is important to note', 'it should be noted', 'it is worth noting',
  'one could argue', 'one might argue',
})

# Define transitional markers frequently used to structure logical arguments
DISCOURSE_MARKERS: frozenset[str] = frozenset({
  'furthermore', 'moreover', 'additionally', 'in conclusion',
  'in summary', 'to summarise', 'firstly', 'secondly', 'thirdly',
  'lastly', 'finally', 'in addition', 'nevertheless', 'nonetheless',
  'consequently', 'therefore', 'thus', 'hence',
})

# Categorise vocabulary often associated with generative AI stylistic patterns
GPTISMS: dict[str, list[str]] = {
  'verbs':      ['delve', 'underscore', 'foster', 'enhance', 'navigate',
                 'utilise', 'leverage', 'facilitate', 'illuminate', 'embody'],
  'adjectives': ['pivotal', 'multifaceted', 'transformative', 'vibrant',
                 'comprehensive', 'nuanced', 'robust', 'dynamic', 'crucial',
                 'intricate', 'profound'],
  'nouns':      ['landscape', 'testament', 'tapestry', 'intersection',
                 'framework', 'paradigm', 'ecosystem', 'dimension', 'realm'],
}

# Flatten the dictionary into a set for efficient membership testing
_ALL_GPTISMS: frozenset[str] = frozenset(
  w for words in GPTISMS.values() for w in words
)


def split_paragraphs(text: str) -> list[str]:
  '''Split on one or more blank lines; discard empty paragraphs.'''
  # Identify paragraph boundaries based on double line breaks
  return [p.strip() for p in re.split(r'\n{2,}', text) if p.strip()]


def tokenise_words(text: str, lowercase: bool = True) -> list[str]:
  '''Word tokens stripped of punctuation; optionally lowercased.'''
  # Convert the raw string into a list of constituent words
  tokens = word_tokenize(text.lower() if lowercase else text)
  # Filter out standalone punctuation marks from the token stream
  return [t for t in tokens if t not in string.punctuation]


def count_phrase_occurrences(text: str, phrases: frozenset[str]) -> int:
  '''
  Count how many times any phrase in *phrases* appears in *text*.
  Multi-word phrases are matched case-insensitively as substrings.
  Single-word phrases benefit from word-boundary anchoring.
  '''
  text_lower = text.lower()
  total = 0
  for phrase in phrases:
    # Use simple count for multi-word strings, regex for individual word accuracy
    if ' ' in phrase:
      total += text_lower.count(phrase)
    else:
      total += len(re.findall(rf'\b{re.escape(phrase)}\b', text_lower))
  return total
````

## Section 2 — Metric Computation

### Mathematical Definitions

1.  **Burstiness**: Defined here as the standard deviation of sentence lengths, $\sigma_{len}$. Human writing typically exhibits [higher variance in sentence structure than AI-generated text](https://arxiv.org/abs/2505.01800v1).
2.  **Type–Token Ratio (TTR)**: A measure of lexical diversity calculated as: $TTR = \frac{\text{Unique Tokens}}{\text{Total Tokens}}$.
3.  **Semantic Redundancy**: Computed via [Cosine Similarity](https://en.wikipedia.org/wiki/Cosine_similarity). For vectors $\mathbf{A}$ and $\mathbf{B}$: $\text{similarity} = \cos(\theta) = \frac{\mathbf{A} \cdot \mathbf{B}}{\|\mathbf{A}\| \|\mathbf{B}\|}$.

````python
# -- 2.1 Syntactic & punctuation -----------------------------------------------

def compute_punctuation_metrics(text: str, metrics: EssayMetrics) -> None:
  '''Mutates *metrics* with colon / semicolon counts and ratio.'''
  metrics.colon_count = text.count(':')
  metrics.semicolon_count = text.count(';')
  # +1 in denominator avoids division-by-zero; ratio > 1 → colon-heavy prose
  metrics.colon_semicolon_ratio = metrics.colon_count / (metrics.semicolon_count + 1)


def compute_sentence_metrics(text: str, metrics: EssayMetrics) -> None:
  '''Mutates *metrics* with per-sentence word counts and derived statistics.'''
  # Segment text into sentences to analyse length variance
  sentences = sent_tokenize(text)
  lengths = [len(word_tokenize(s)) for s in sentences]
  metrics.sentence_lengths = lengths
  if lengths:
    # Calculate central tendency and dispersion of sentence lengths
    metrics.mean_sentence_length = float(np.mean(lengths))
    metrics.sentence_length_std  = float(np.std(lengths))


def compute_hedge_metrics(text: str, metrics: EssayMetrics, word_count: int) -> None:
  '''Mutates *metrics* with hedge-word and discourse-marker rates.'''
  # Quantify the presence of specific rhetorical markers
  hedge_hits    = count_phrase_occurrences(text, HEDGE_WORDS)
  discourse_hits = count_phrase_occurrences(text, DISCOURSE_MARKERS)
  # Normalise counts to a rate per 100 words for comparability across lengths
  per_100 = 100 / max(word_count, 1)
  metrics.hedge_frequency         = hedge_hits    * per_100
  metrics.discourse_marker_frequency = discourse_hits * per_100


# -- 2.2 Lexical ---------------------------------------------------------------

def compute_lexical_metrics(
  tokens: list[str],
  text: str,
  metrics: EssayMetrics,
) -> None:
  '''Mutates *metrics* with TTR and GPT-ism statistics.'''
  # Measure lexical variety by the ratio of unique to total words
  if tokens:
    metrics.ttr = len(set(tokens)) / len(tokens)

  gptism_counts: dict[str, int] = {}
  total_gptisms = 0
  # Identify individual occurrences of high-probability AI vocabulary
  for word in _ALL_GPTISMS:
    n = len(re.findall(rf'\b{re.escape(word)}\b', text.lower()))
    if n:
      gptism_counts[word] = n
      total_gptisms += n

  metrics.gptism_counts     = gptism_counts
  metrics.gptism_frequency  = total_gptisms * (100 / max(len(tokens), 1))


# -- 2.3 Structural -----------------------------------------------------------

def compute_structural_metrics(
  text: str,
  paragraphs: list[str],
  metrics: EssayMetrics,
  word_count: int,
) -> None:
  '''Mutates *metrics* with paragraph-level structural signals.'''
  # Track the word counts within individual paragraphs
  para_lengths = [len(word_tokenize(p)) for p in paragraphs]
  metrics.paragraph_lengths    = para_lengths
  metrics.paragraph_length_std = float(np.std(para_lengths)) if para_lengths else 0.0

  # Signposting: discourse markers that open sentences
  # Count occurrences of transitional phrases at a global level
  signpost_hits = count_phrase_occurrences(text, DISCOURSE_MARKERS)
  metrics.signpost_frequency = signpost_hits * (100 / max(word_count, 1))

  # 'Nuance trap': concessive constructions — 'while ... , ... ' / 'although ... '
  # Identify concessive conjunctions often over-used in structured AI summaries
  nuance_pattern = re.compile(
    r'\b(while|although|even though|despite|whereas)\b',
    re.IGNORECASE,
  )
  metrics.nuance_trap_frequency = (
    len(nuance_pattern.findall(text)) * (100 / max(word_count, 1))
  )

  # Intro / outro similarity — compare first and last paragraph embeddings
  # Evaluate semantic similarity between introduction and conclusion
  if len(paragraphs) >= 2:
    metrics.intro_outro_similarity = _paragraph_cosine_similarity(
      paragraphs[0], paragraphs[-1]
    )


def _paragraph_cosine_similarity(para_a: str, para_b: str) -> float:
  '''TF-IDF cosine similarity between two text blocks.'''
  # Transform text into sparse frequency vectors to compute alignment
  vectoriser = TfidfVectorizer().fit([para_a, para_b])
  vecs = vectoriser.transform([para_a, para_b])
  return float(cosine_similarity(vecs[0], vecs[1])[0, 0])


# -- 2.4 Sophisticated / semantic ---------------------------------------------

def compute_semantic_redundancy(paragraphs: list[str], metrics: EssayMetrics) -> None:
  '''
  Measures intra-paragraph semantic redundancy via mean pairwise sentence
  cosine similarity within each paragraph.  High values → AI-style rewording.
  '''
  all_sims: list[float] = []

  for para in paragraphs:
    # Break paragraph into sentences for internal similarity checks
    sents = sent_tokenize(para)
    if len(sents) < 2:
      continue
    try:
      # Construct a local vectoriser for the sentences in this specific paragraph
      vectoriser = TfidfVectorizer().fit(sents)
      vecs = vectoriser.transform(sents).toarray()
    except ValueError:
      # All sentences identical or empty after TF-IDF pruning
      # Handle cases where sentences contain only stop-words or identical content
      continue

    # Upper-triangle pairwise similarities (excluding self-similarity)
    # Calculate the average similarity between every pair of sentences
    n = len(sents)
    sims = [
      float(1 - cosine(vecs[i], vecs[j]))
      for i in range(n)
      for j in range(i + 1, n)
      if np.any(vecs[i]) and np.any(vecs[j])   # skip zero vectors
    ]
    if sims:
      all_sims.append(float(np.mean(sims)))

  metrics.intra_paragraph_similarities = all_sims
  metrics.mean_intra_paragraph_similarity = (
    float(np.mean(all_sims)) if all_sims else 0.0
  )
````

## Section 3 — Orchestrator

### Pipeline Execution

The orchestrator functions sequence the modular metric computations. `analyse_essay` transforms raw strings into the `EssayMetrics` schema, while `metrics_to_series` acts as a bridge to `pandas`, enabling efficient corpus-level aggregation and manipulation.

````python
def analyse_essay(text: str, label: str = 'essay') -> EssayMetrics:
  '''
  Full pipeline: run all metric groups and return a populated EssayMetrics.
  *label* is a human-readable identifier (e.g. 'pre-ChatGPT', 'Student A').
  '''
  metrics    = EssayMetrics(label=label)
  paragraphs = split_paragraphs(text)
  tokens     = tokenise_words(text)
  word_count = len(tokens)

  compute_punctuation_metrics(text, metrics)
  compute_sentence_metrics(text, metrics)
  compute_hedge_metrics(text, metrics, word_count)
  compute_lexical_metrics(tokens, text, metrics)
  compute_structural_metrics(text, paragraphs, metrics, word_count)
  compute_semantic_redundancy(paragraphs, metrics)

  return metrics


def metrics_to_series(m: EssayMetrics) -> pd.Series:
  '''Flatten scalar fields of EssayMetrics to a labelled pandas Series.'''
  return pd.Series({
    'label':                          m.label,
    'colon_semicolon_ratio':          m.colon_semicolon_ratio,
    'mean_sentence_length':           m.mean_sentence_length,
    'sentence_length_std':            m.sentence_length_std,
    'hedge_frequency':                m.hedge_frequency,
    'discourse_marker_frequency':     m.discourse_marker_frequency,
    'ttr':                            m.ttr,
    'gptism_frequency':               m.gptism_frequency,
    'paragraph_length_std':           m.paragraph_length_std,
    'signpost_frequency':             m.signpost_frequency,
    'nuance_trap_frequency':          m.nuance_trap_frequency,
    'intro_outro_similarity':         m.intro_outro_similarity,
    'mean_intra_para_similarity':     m.mean_intra_paragraph_similarity,
  })
````

## Section 4 — Pairwise Comparison

### Statistical Evaluation

We compare corpora using the [Mann–Whitney U test](https://en.wikipedia.org/wiki/Mann%E2%80%93Whitney_U_test). This non-parametric test evaluates the null hypothesis that it is equally likely that a randomly selected value from one sample will be less than or greater than a randomly selected value from a second sample. It is particularly robust against the skewed distributions common in linguistic data.

````python
def compare_texts(
  text1: str,
  text2: str,
  label1: str = 'Text 1',
  label2: str = 'Text 2',
) -> tuple[EssayMetrics, EssayMetrics, pd.DataFrame]:
  '''
  Analyse two essays and return both EssayMetrics objects and a side-by-side
  comparison DataFrame with an absolute-delta column.
  '''
  m1 = analyse_essay(text1, label=label1)
  m2 = analyse_essay(text2, label=label2)

  s1 = metrics_to_series(m1)
  s2 = metrics_to_series(m2)

  numeric_keys = [k for k in s1.index if k != 'label']

  comparison = pd.DataFrame({
    label1: s1[numeric_keys].astype(float),
    label2: s2[numeric_keys].astype(float),
  })
  comparison['Δ (2 − 1)'] = comparison[label2] - comparison[label1]
  comparison['Δ %']       = (
    comparison['Δ (2 − 1)'] / comparison[label1].replace(0, np.nan) * 100
  ).round(1)

  return m1, m2, comparison


def compare_corpora(
  corpus1: list[str],
  corpus2: list[str],
  label1: str = 'Corpus 1',
  label2: str = 'Corpus 2',
) -> tuple[pd.DataFrame, pd.DataFrame]:
  '''
  Analyse n-vs-n corpora; return per-essay DataFrame and a summary
  (mean ± SD per group) suitable for group-level visualisation.

  Uses Mann–Whitney U for non-parametric significance testing.
  '''
  rows1 = [metrics_to_series(analyse_essay(t, label=label1)) for t in corpus1]
  rows2 = [metrics_to_series(analyse_essay(t, label=label2)) for t in corpus2]

  df = pd.DataFrame(rows1 + rows2)
  numeric_cols = df.columns.difference(['label'])
  df[numeric_cols] = df[numeric_cols].astype(float)

  # -- Per-group summary with significance -----------------------------------
  summary_rows = []
  for col in numeric_cols:
    g1 = df.loc[df['label'] == label1, col].dropna()
    g2 = df.loc[df['label'] == label2, col].dropna()
    stat, p = mannwhitneyu(g1, g2, alternative='two-sided') if (len(g1) > 1 and len(g2) > 1) else (np.nan, np.nan)
    summary_rows.append({
      'metric':              col,
      f'{label1}_mean':      g1.mean(),
      f'{label1}_std':       g1.std(),
      f'{label2}_mean':      g2.mean(),
      f'{label2}_std':       g2.std(),
      'mann_whitney_u':      stat,
      'p_value':             p,
      'significant_p05':     p < 0.05 if not np.isnan(p) else False,
    })

  summary = pd.DataFrame(summary_rows).set_index('metric')
  return df, summary
````

The 'Corpus Summary' includes error bars representing one standard deviation and significance markers (★) for metrics where $p < 0.05$. This provides a 'at-a-glance' diagnostic of which linguistic features have shifted most significantly over time.

````python
# -- Shared helpers ------------------------------------------------------------

def _bar_comparison(
  ax: plt.Axes,
  labels: list[str],
  values: list[float],
  metric_name: str,
  colours: Optional[list[str]] = None,
) -> None:
  colours = colours or _PALETTE_TWO
  bars = ax.bar(labels, values, color=colours[:len(labels)], width=0.5)
  ax.bar_label(bars, fmt='%.2f', padding=3, fontsize=9)
  ax.set_title(metric_name.replace('_', ' ').title(), fontsize=10)
  ax.set_ylabel('Score')


def _sentence_length_histogram(
  ax: plt.Axes,
  m1: EssayMetrics,
  m2: EssayMetrics,
) -> None:
  bins = np.linspace(0, max(max(m1.sentence_lengths, default=1),
                             max(m2.sentence_lengths, default=1)) + 5, 20)
  ax.hist(m1.sentence_lengths, bins=bins, alpha=0.7, label=m1.label, color=_PALETTE_TWO[0])
  ax.hist(m2.sentence_lengths, bins=bins, alpha=0.7, label=m2.label, color=_PALETTE_TWO[1])
  ax.set_title('Sentence Length Distribution (Burstiness)')
  ax.set_xlabel('Words per sentence')
  ax.set_ylabel('Count')
  ax.legend()


# -- Pairwise dashboard --------------------------------------------------------

_PAIRWISE_SCALAR_METRICS: list[str] = [
  'colon_semicolon_ratio',
  'mean_sentence_length',
  'sentence_length_std',
  'hedge_frequency',
  'discourse_marker_frequency',
  'ttr',
  'gptism_frequency',
  'paragraph_length_std',
  'intro_outro_similarity',
  'mean_intra_para_similarity',
]


def plot_pairwise(
  m1: EssayMetrics,
  m2: EssayMetrics,
  figsize: tuple[float, float] = (18, 14),
) -> plt.Figure:
  '''
  Produce a multi-panel dashboard comparing two EssayMetrics objects.
  Returns the Figure so callers can save or further customise it.
  '''
  n_scalar = len(_PAIRWISE_SCALAR_METRICS)
  # Layout: scalar bars (2 cols) + histogram + GPT-ism breakdown
  n_bar_rows = (n_scalar + 1) // 2
  total_rows  = n_bar_rows + 2   # extra rows for histogram & GPT-ism chart

  fig = plt.figure(figsize=figsize, constrained_layout=True)
  fig.suptitle(
    f'Essay Metric Comparison: {m1.label} vs {m2.label}',
    fontsize=14, fontweight='bold',
  )
  gs = gridspec.GridSpec(total_rows, 2, figure=fig)

  # -- Scalar bar charts -----------------------------------------------------
  for idx, metric in enumerate(_PAIRWISE_SCALAR_METRICS):
    row, col = divmod(idx, 2)
    ax = fig.add_subplot(gs[row, col])
    _bar_comparison(
      ax,
      labels=[m1.label, m2.label],
      values=[getattr(m1, metric), getattr(m2, metric)],
      metric_name=metric,
    )

  # -- Sentence length histogram ---------------------------------------------
  ax_hist = fig.add_subplot(gs[n_bar_rows, :])
  _sentence_length_histogram(ax_hist, m1, m2)

  # -- GPT-ism breakdown (grouped bar) --------------------------------------
  ax_gpt = fig.add_subplot(gs[n_bar_rows + 1, :])
  _plot_gptism_breakdown(ax_gpt, m1, m2)

  return fig


def _plot_gptism_breakdown(
  ax: plt.Axes,
  m1: EssayMetrics,
  m2: EssayMetrics,
) -> None:
  '''Grouped bar chart of individual GPT-ism word counts.'''
  all_words = sorted(set(m1.gptism_counts) | set(m2.gptism_counts))
  if not all_words:
    ax.set_visible(False)
    return

  x      = np.arange(len(all_words))
  width  = 0.35
  counts1 = [m1.gptism_counts.get(w, 0) for w in all_words]
  counts2 = [m2.gptism_counts.get(w, 0) for w in all_words]

  ax.bar(x - width / 2, counts1, width, label=m1.label, color=_PALETTE_TWO[0])
  ax.bar(x + width / 2, counts2, width, label=m2.label, color=_PALETTE_TWO[1])
  ax.set_xticks(x)
  ax.set_xticklabels(all_words, rotation=35, ha='right', fontsize=9)
  ax.set_title('GPT-ism Word Counts')
  ax.set_ylabel('Count')
  ax.legend()


# -- Corpus-level visualisation ------------------------------------------------

def plot_corpus_summary(
  summary: pd.DataFrame,
  label1: str = 'Corpus 1',
  label2: str = 'Corpus 2',
  figsize: tuple[float, float] = (14, 10),
) -> plt.Figure:
  '''
  Horizontal bar chart of per-group means with ± 1 SD error bars.
  Significant metrics (p < .05) are marked with *.
  '''
  metrics = summary.index.tolist()
  means1  = summary[f'{label1}_mean'].values
  means2  = summary[f'{label2}_mean'].values
  stds1   = summary[f'{label1}_std'].values
  stds2   = summary[f'{label2}_std'].values
  sig     = summary['significant_p05'].values

  y     = np.arange(len(metrics))
  height = 0.35

  fig, ax = plt.subplots(figsize=figsize, constrained_layout=True)
  ax.barh(y + height / 2, means1, height, xerr=stds1, label=label1,
          color=_PALETTE_TWO[0], alpha=0.9, capsize=3)
  ax.barh(y - height / 2, means2, height, xerr=stds2, label=label2,
          color=_PALETTE_TWO[1], alpha=0.9, capsize=3)

  # Mark statistically significant metrics
  for i, (metric, is_sig) in enumerate(zip(metrics, sig)):
    if is_sig:
      ax.text(
        max(means1[i] + stds1[i], means2[i] + stds2[i]) * 1.02,
        i,
        '★ p<.05',
        va='center', fontsize=8, color='#c00000',
      )

  ax.set_yticks(y)
  ax.set_yticklabels([m.replace('_', ' ').title() for m in metrics])
  ax.set_xlabel('Mean score')
  ax.set_title(f'Corpus Metric Comparison: {label1} vs {label2}', fontweight='bold')
  ax.legend()
  ax.invert_yaxis()

  return fig


def plot_sentence_burstiness_corpora(
  df: pd.DataFrame,
  metric: str = 'sentence_length_std',
  label1: str = 'Corpus 1',
  label2: str = 'Corpus 2',
  figsize: tuple[float, float] = (8, 5),
) -> plt.Figure:
  '''Box + strip plot for any metric across two corpora.'''
  fig, ax = plt.subplots(figsize=figsize, constrained_layout=True)
  sns.boxplot(data=df, x='label', y=metric, ax=ax, order=[label1, label2],
              palette=_PALETTE_TWO, width=0.4)
  sns.stripplot(data=df, x='label', y=metric, ax=ax, order=[label1, label2],
                color='black', alpha=0.5, size=4, jitter=True)
  ax.set_title(f'{metric.replace("_", " ").title()} by Corpus', fontweight='bold')
  ax.set_xlabel('')
  ax.set_ylabel(metric.replace('_', ' ').title())
  return fig
````

## Design Notes

| Decision | Rationale |
|---|---|
| `@dataclass` for `EssayMetrics` | Typed, auto-`__repr__`, trivially extensible — add a field and the pipeline picks it up |
| Mutating helpers (`compute_*`) | Keeps each concern isolated; avoids re-passing the same object as both input and return |
| `metrics_to_series` bridge | Single conversion point; keeps corpus logic agnostic of the dataclass internals |
| `+1` denominator in ratios | Avoids silent `ZeroDivisionError` on short texts — preserves expected direction of the metric |
| Mann–Whitney U (not t-test) | Essay scores are rarely normal; non-parametric test is safer for small, skewed corpora |

**Next natural extension points:**
- Add a `stylometric_delta` module using Burrows' Delta across an *n*-way corpus
- Swap TF-IDF paragraph embeddings for sentence-transformers (`all-MiniLM-L6-v2`) once you want semantic rather than lexical similarity
- Wrap `analyse_essay` in a file-loader that accepts `.txt` / `.docx` so the notebook cell stays clean
